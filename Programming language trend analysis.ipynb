```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("QueryResults.csv", names= ['DATE','TAG','POSTS'],header=0)
data.head()
```

```python
type(data['DATE'][2])
```

```python
data['DATE'] = pd.to_datetime(data['DATE'])
data.head()
```

```python
reshaped_data = data.pivot(index='DATE', columns='TAG', values='POSTS')
reshaped_data
```

```python
reshaped_data.shape
```

```python
reshaped_data.count()
```

```python
reshaped_data.fillna(0, inplace = True)
```

```python
reshaped_data
```

```python
reshaped_data.isna()
```

```python
reshaped_data.isna().values.any()
```

```python
# To know the popularity of java programming language
plt.figure(figsize=(16,10))
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0, 35000)
plt.plot(reshaped_data.index, reshaped_data.java)
```

```python
plt.figure(figsize=(8,5))
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0, 35000)
plt.plot(reshaped_data.index, reshaped_data.go)
```

```python
# to compare trend between python and C language
plt.figure(figsize=(8,5))
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0, 35000)
plt.plot(reshaped_data.index, reshaped_data.python)
plt.plot(reshaped_data.index, reshaped_data.c)
```

```python
plt.figure(figsize=(20,15))
for columns in reshaped_data:
    plt.plot(reshaped_data.index, reshaped_data[columns], linewidth = 3, label=reshaped_data[columns].name)
    plt.legend(fontsize=16)
```

```python
roll_df = reshaped_data.rolling(window=12). mean()

plt.figure(figsize=(16,10))
plt.xticks(fontsize = 14)
plt.yticks(fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0, 35000)
for columns in reshaped_data:
    plt.plot(roll_df.index, roll_df[columns], linewidth = 3, label=roll_df[columns].name)
plt.legend(fontsize=16)
```
